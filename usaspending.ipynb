```python
# import dependencies 
import requests
import pandas as pd
```

```python
# access active 8a firms list 
datafile = 'SBA 8a Active Firms with Recip ID.csv'
```

```python
# create dataframe
usa_df = pd.read_csv(datafile)

# change DUNS number to string 
usa_df["DUNS_Number"] = usa_df["DUNS_Number"].astype(str)
```

```python
# use zfill so the leading 0's in duns numbers dont get lost 
usa_df["DUNS_Number"] = usa_df["DUNS_Number"].str.zfill(9)
usa_df
```

```python
# create lists to use later 
recipient_list = []
recip_id_list = []
```

```python
# pull duns column from data frame to use as list in for loop 
duns_list = usa_df["DUNS_Number"].astype(str).str.zfill(9).tolist()
```

```python
# this takes an hour
for i in duns_list:
    url = "https://api.usaspending.gov/api/v2/recipient/children/"
    duns = i
    query_url = url + duns
    print(f"processing {duns}...")
    
    try:
        spending_data = requests.get(query_url)
        response_data = spending_data.json()
        recip_id = response_data[0]["recipient_id"]
        name = response_data[0]["name"]
        duns = response_data[0]["duns"]
        recipient_list.append({"Recipient ID":recip_id,
                              "Name":name,
                              "DUNS":duns})
    except:
        print("duns not found. skipping...")
        pass
```

```python
recipient_list
```

```python
final_df = pd.DataFrame(recipient_list, columns = ['Recipient ID', 'Name', 'DUNS'])
final_df.head(20)
```

```python
final_df.to_csv('recipient_id.csv')
```

```python
merged_df = pd.merge(usa_df,final_df,left_on="DUNS_Number",right_on="DUNS",how="left")
merged_df.to_csv('merged_df.csv')
merged_df
```

```python
# to get NaN values: 
# 1) use DUNS to export transaction data, 
# 2) get parent duns, 
#  3) re-run api, 
#  4) and search for child with match DUNS
```

```python
recipient_ids = merged_df["Recipient ID"].astype(str).tolist()
```

```python
for j in recipient_ids:
    url = "https://api.usaspending.gov/api/v2/recipient/duns/"
    recipID = j
    query_url = url + recipID
    print(f"processing {recipID}...")

    try:
        recipient_data = requests.get(query_url)
        response_data = recipient_data.json()
        name = response_data["name"]
        duns = response_data["duns"]
        business_types = response_data["business_types"]
        total_transaction_amount = response_data["total_transaction_amount"]
        total_transactions = response_data["total_transactions"]
        recipient_list.append({"Name":name,
                               "DUNS":duns,
                               "Business Types":business_types,
                               "Total Transaction Amount":total_transaction_amount,
                              "Total Transactions":total_transactions})
    
    except:
        print(f"{recipID} not found. skipping...") 
        pass
```

```python
df = pd.DataFrame(recipient_list, columns = ['Name','DUNS','Business Types','Total Transaction Amount','Total Transactions'])
df
```

```python
new_merged_df = pd.merge(merged_df,df,left_on="DUNS_Number",right_on="DUNS",how="left")
new_merged_df
```

```python
new_merged_df.to_csv('new_merged_df.csv')
```
