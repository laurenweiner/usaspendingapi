```python
# import dependencies
import requests
import pandas as pd
import pprint
```

```python
# clear list 
award_list=[]
award_list.clear()
```

```python
# access active 8a firmst list 
datafile = 'SBA 8a Active Firms with Recip ID.csv'

# create dataframe
usa_df = pd.read_csv(datafile)

# change DUNS number to string 
#usa_df["DUNS_Number"] = usa_df["DUNS_Number"].astype(str)

# pull duns column from data frame to use as list in for loop 
duns_list = usa_df["DUNS"].astype(str).str.zfill(9).tolist()
```

```python
num_list = list(range(1,20000))
```

```python
# spending_by_award
award_endpoint = 'https://api.usaspending.gov/api/v2/search/spending_by_award/'
```

```python
for i in num_list: 

    # filters
    data =     {
          "subawards": 'false',
          "limit": 100,
          "page": i,
          "filters": {
              "award_type_codes": ["A", "B", "C","D"],
              "recipient_search_text": duns_list[0:500]
          },
          "fields": [
              "Award ID",
              "generated_internal_id",
              "Recipient Name",
              "Recipient DUNS Number",
              "Base Obligation Date",
              "Start Date",
              "End Date",
              "Place of Performance State Code",
              "Award Amount",
              "Awarding Agency Code",
              "Awarding Sub Agency Code",
              "Contract Award Type",
              "Award Type",
              "Funding Agency Code",
              "Funding Sub Agency Code",
              "COVID-19 Outlays",
              "COVID-19 Obligations"
          ]
      }
    
    print(f'processing page {i}')

    # post request 
    k = requests.post(url = award_endpoint, json = data)
    response_k = k.json()
    results_k = response_k['results']
    

    # go through list of awards and organize 
    for award in results_k:
        award_id = award["Award ID"]
        generated_internal_id = award["generated_internal_id"]
        recipient_name = award["Recipient Name"]
        recipient_duns = award["Recipient DUNS Number"]
        base_obligation_date = award["Base Obligation Date"]
        start_date = award["Start Date"]
        end_date = award["End Date"]
        pop_state_code = award["Place of Performance State Code"]
        award_amount = award["Award Amount"]
        awarding_agency_code = award["Awarding Agency Code"]
        awarding_subagency_code = award["Awarding Sub Agency Code"]
        contract_award_type = award["Contract Award Type"]
        award_type = award["Award Type"]
        funding_agency_code = award["Funding Agency Code"]
        funding_subagency_code = award["Funding Sub Agency Code"]
        covid_outlays = award["COVID-19 Outlays"]
        covid_obligations = award["COVID-19 Obligations"]
        award_list.append({'Award ID':award_id,
                           'generated_internal_id':generated_internal_id,
                           'Recipient Name':recipient_name, 
                           'Recipient DUNS Number':recipient_duns,
                           'Base Obligation Date':base_obligation_date,
                          'Start Date':start_date,
                          'End Date':end_date,
                           'Place of Performance State Code':pop_state_code,
                          'Award Amount':award_amount,
                          'Awarding Agency Code':awarding_agency_code,
                          'Awarding Sub Agency Code':awarding_subagency_code,
                          'Contract Award Type':contract_award_type,
                          'Award Type':award_type,
                          'Funding Agency Code':funding_agency_code,
                          'Funding Sub Agency Code':funding_subagency_code,
                           'COVID-19 Outlays':covid_outlays,
                           'COVID-19 Obligations':covid_obligations
                          })
            
    if not results_k:
        print(f'page {i} is empty')
        break
```
