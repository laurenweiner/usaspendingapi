```python
# import dependencies
import requests
import pandas as pd
import pprint
```

```python
# clear list 
award_list=[]
award_list.clear()
```

```python
# access active 8a firmst list 
datafile = 'SBA 8a Active Firms with Recip ID.csv'

# create dataframe
usa_df = pd.read_csv(datafile)

# change DUNS number to string 
#usa_df["DUNS_Number"] = usa_df["DUNS_Number"].astype(str)

# pull duns column from data frame to use as list in for loop 
duns_list = usa_df["DUNS"].astype(str).str.zfill(9).tolist()
```

```python
num_list = list(range(1,20000))
```

```python
# spending_by_award
award_endpoint = 'https://api.usaspending.gov/api/v2/search/spending_by_award/'
```

```python
for i in num_list: 

    # filters
    data =     {
          "subawards": 'false',
          "limit": 100,
          "page": i,
          "filters": {
              "award_type_codes": ["A", "B", "C","D"],
              "recipient_search_text": duns_list[1001:2000]
          },
          "fields": [
              "Award ID",
              "Recipient Name",
              "Recipient DUNS Number",
              "Start Date",
              "End Date",
              "Award Amount",
              "Awarding Agency",
              "Awarding Sub Agency",
              "Contract Award Type",
              "Award Type",
              "Funding Agency",
              "Funding Sub Agency"
          ]
      }
    
    print(f'processing page {i}')

    # post request 
    k = requests.post(url = award_endpoint, json = data)
    response_k = k.json()
    results_k = response_k['results']
    

    # go through list of awards and organize 
    for award in results_k:
        award_id = award["Award ID"]
        recipient_name = award["Recipient Name"]
        recipient_duns = award["Recipient DUNS Number"]
        start_date = award["Start Date"]
        end_date = award["End Date"]
        award_amount = award["Award Amount"]
        awarding_agency = award["Awarding Agency"]
        awarding_subagency = award["Awarding Sub Agency"]
        contract_award_type = award["Contract Award Type"]
        award_type = award["Award Type"]
        funding_agency = award["Funding Agency"]
        funding_subagency = award["Funding Sub Agency"]
        award_list.append({'Award ID':award_id,
                           'Recipient Name':recipient_name, 
                           'Recipient DUNS Number':recipient_duns,
                          'Start Date':start_date,
                          'End Date':end_date,
                          'Award Amount':award_amount,
                          'Awarding Agency':awarding_agency,
                          'Awarding Sub Agency':awarding_subagency,
                          'Contract Award Type':contract_award_type,
                          'Award Type':award_type,
                          'Funding Agency':funding_agency,
                          'Funding Sub Agency':funding_subagency
                          })
            
    if not results_k:
        print(f'page {i} is empty')
        break
```

```python
# make data frame to view data easily 
awards_df = pd.DataFrame(award_list, columns = ['Award ID',
                                                'Recipient Name',
                                                'Recipient DUNS Number',
                                                'Start Date','End Date',
                                                'Award Amount', 
                                                'Awarding Agency',
                                                'Awarding Sub Agency',
                                                'Contract Award Type',
                                                'Award Type',
                                                'Funding Agency',
                                                'Funding Sub Agency'])
awards_df
```
